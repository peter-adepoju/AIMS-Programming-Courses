(*) Compute the conductor and Frobenius number of the following pairs: $\{4,7\}$, $\{4,6\}$, and $\{7,9\}$, if they exist. How many numbers are not writable in each case?

```python
"""
I used the function provided in the day 03 lecture notebook 
"""
def possible_cents(a , b, bound):
    # Initialise all_possible_cents to be the empty list
    all_possible_cents = list() #  all_possible_cents = []
    for x in [0..bound]:
        for y in [0..bound]:
            all_possible_cents.append(a * x + b * y)
    return sorted(all_possible_cents) # all_possible_cents.sort() this will return None.
```

```python
def find_gaps(a, b, bound):
    r"""
    Finds and returns the missing numbers (gaps) in the sequence until there are n consecutive numbers without gaps.
    """
    n = abs(a - b) + 1  # Define n as |a - b| + 1
    sequence = possible_cents(a, b, bound)  # Generate the sequence
    
    missing_values = []  # List to store missing values in gaps
    max_value = max(sequence)  # Find the maximum value in the sequence
    all_numbers = list(range(max_value + 1))  # Full range of numbers up to max_value
    
    gap_counter = 0  # Counter to track consecutive numbers without gaps
    for number in all_numbers:        # Loop through the full range and check for gaps
        if number in sequence:
            gap_counter += 1  # Increment if a number is in the sequence
            if gap_counter == n:  # Stop if we reach n consecutive numbers without a gap
                break
        else:
            missing_values.append(number)  # Add to missing values if it's a gap
            gap_counter = 0  # Reset counter if we find a gap
   
    return missing_values
```

  1. Suppose $A \subseteq C$, and $B$ and $C$ are disjoint. Prove that if $x \in A$ then $x \notin B$.

#### Formal proof of Statement 1:

Suppose $A \subseteq C$, and $B$ and $C$ are disjoint. Prove that if $x \in A$ then $x \notin B$.

Assume $x \in A$. Since $A \subseteq C$, it follows that $x \in C$ (because all elements of $A$ are also elements of $C$). Since $x \in C$ and $B \cap C = \emptyset$, this implies that $x \notin B$ 

#### Formal proof of statement 2: 

Suppose $A \subseteq C$, and $B$ and $C$ are disjoint. Prove that if $x \in A$ then $x \notin B$.

Assume $x \in A$ and $x \in C$. Since $ (A \setminus B) \cap C = \emptyset$, this means there is no element in both $A \setminus B$ and $C$. Given that $x \in A$ and $x \in C$, if $x$ were in $B$, then it cannot be in $A \setminus B$. Thus, $x$ cannot be in $A \setminus B$ because that would contradict $x \in C$. Therefore, it must be that $x \in B$.

#### Formal proof of statement 3a:
Suppose $A \cap C \subseteq B$ and $a \in C$. Prove that $a \notin A \setminus B$.

**Proof by Contradiction:** Assume for the sake of contradiction that $a \in A \setminus B$, this means that $a \in A$ but $a \notin B$. Based on our assumption that $a \in C$, it follows that $a \in A \cap C$. However, we have $A \cap C \subseteq B$, which means that every element in $A \cap C$ must 
also be in $B$. This leads to a **contradiction** since $a \notin B$.

#### Formal proof of statement 3b:
Suppose that $A \subseteq B$, $a\in A$, and $a \notin B \setminus C$. Prove that $a \in C$.

**Proof by Contradiction:** Assume for the sake of contradiction that $a \notin C$. Since $a \in A$ and $A \subseteq B$, it follows that $a \in B$. However, by assumption $a \notin B \setminus C$, this implies that if $a \in B$, it must be that $a \in C$ (because $B \setminus C$ contains elements that are in $B$ but not in $C$). This is a **contradiction**, since we assumed $a \notin C$.
