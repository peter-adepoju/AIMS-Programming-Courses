```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import polyfit

# Example data for r and t
t = np.array([5, 15, 30, 45, 60, 90, 120, 180, 240, 2400])  # Time in seconds
r = np.array([8.0, 9.650, 10.225, 10.600, 10.950, 11.425, 11.775, 12.125, 12.550, 15.6])  # Radius in cm

# Take logarithms
log_t = np.log(t)
log_r = np.log(r)

# Linear regression
b, log_a = np.polyfit(log_t, log_r, 1) # 1 indicates a linear fit
a = np.exp(log_a)  # Assuming natural log was used

print(f"a = {a:.2f}")
print(f"b = {b:.2f}")

# Plot log(r) against log(t)
plt.figure(figsize=(8, 6))
plt.plot(log_t, log_r, marker='o', linestyle='-', color='b', label='ln r_avg (cm) vs log t(s)')
plt.xlabel('ln t (s)', fontsize=12)
plt.ylabel('ln r_avg (cm)', fontsize=12)
plt.title('Plot of ln r_avg (cm) against ln t (s)', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data for r and t
t = np.array([5, 15, 30, 45, 60, 90, 120, 180, 240, 2400])  # Time in seconds
r = np.array([8.0, 9.650, 10.225, 10.600, 10.950, 11.425, 11.775, 12.125, 12.550, 15.6])  # Radius in cm

# Take logarithms
log_t = np.log(t)
log_r = np.log(r)

# Linear regression
b, log_a = np.polyfit(log_t, log_r, 1) # 1 indicates a linear fit
a = np.exp(log_a)  # Assuming natural log was used

print(f"a = {a:.3f}")
print(f"b = {b:.3f}")


# --- Plotting ---
plt.figure(figsize=(8, 6)) # size of the plot
plt.plot(log_t, log_r, marker='o', linestyle='-', color='#2A367C', markersize = 5,  label='Experimental Data')  # Color, marker size

# Add a regression line
line_x = np.linspace(min(log_t), max(log_t), 500)  
line_y = np.polyval([b, log_a], line_x)  
plt.plot(line_x, line_y, linestyle='--', color='#E93B81', linewidth=1.5, label=f'Linear Fit,  y = {b:.3f} x + {log_a:.3f}')

# axis labels, using appropriate units
plt.xlabel(r'$\ln t \ (s)$', fontsize=12)
plt.ylabel(r'$\ln r_{\text{avg}} \ (cm)$', fontsize=12)

# Title with fontsize
plt.title('Log-Log Plot of Radius vs. Time for Spreading Syrup', fontsize=14)

# add the grid
plt.grid(True, linestyle='--', alpha=0.7)

# add the legend
plt.legend(fontsize=10) # adding legend and fontsize

# adding space around the plot
plt.margins(0.05)

# tight layout to remove any gaps
plt.tight_layout()


plt.show()
```
